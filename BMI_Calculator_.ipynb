The Body Mass Index (BMI) Calculator can be used to calculate BMI value and corresponding weight status while taking age into consideration.BMI, more often than not, correlates with total body fat. This means that as the BMI score increases, so does a person’s total body fat.

  The formula is - BMI = (Weight in kilograms) divided by (Height in metres squared).

 A normal BMI score is one that falls between 18.5 and 24.9. This indicates that a person is within the normal weight range for his or her height.

So lets go and see in which catogary we fall according to our weight and height.


```python
# First creat a function name bmi_calculator.
def bmi_calculator():
    height = float(input("Enter your height in centimeters (cms) "))                 #user input 

    weight = float(input("Enter your weight in kg "))                                          # user input

    height = height / 100

    BMI = weight / (height ** 2)                                               # Formulae for bmi calculator

    print("Your Body Mass Index is ",BMI)                             #output to user

# condition 1 if BMI is 0 and less than or equal to 16.
    if BMI > 0:                                                  
        if BMI <= 16:
            print("You are severley underweight")                        #output to user     

# Condition 3 if BMI is over 16 but under or 18
        elif BMI <= 18:
            print("You are underweight")                                 #output to user

# Condition 3 if BMI is over 18 but under or 25
        elif BMI <= 25:
            print("You are Healthy")                                      #output to user

# Condition 4 if BMI is over 25 but under or  30
        elif BMI <= 30:
            print("You are overweight")                                     #output to user

        else:
            print("You are severley overweight")                           #output to user

# Condition 5 if BMI is over 30 
    else:
        print("Please enter correct values")
    return BMI
```
